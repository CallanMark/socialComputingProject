```python
# order by test accuracy
politifact_results = politifact_results.sort_values(by='test_accuracy', ascending=False)
politifact_results.iloc[0]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm

# Prepare data (drop rows with missing test_accuracy)
plot_data = politifact_results.dropna(subset=['test_accuracy'])

# =========================
# Define Hyperparameter Types
# =========================
continuous_hyperparameters = ['dropout', 'lr', 'weight_decay']  # TRUE continuous only
categorical_hyperparameters = [
    'batch_size', 'heads', 'hidden_dim', 'num_layers',
    'ensemble_method', 'pooling', 'use_self_loops'
]

# =========================
# Select Top 10% and Bottom 10% Runs
# =========================
top_percent = 0.10
n_samples = int(top_percent * len(plot_data))

top_10pct = plot_data.sort_values(by='test_accuracy', ascending=False).head(n_samples)
bottom_10pct = plot_data.sort_values(by='test_accuracy', ascending=True).head(n_samples)

print(f"\nTop {top_percent*100:.0f}% Test Accuracies:")
display(top_10pct[['test_accuracy']])

print(f"\nBottom {top_percent*100:.0f}% Test Accuracies:")
display(bottom_10pct[['test_accuracy']])

# =========================
# Categorical Hyperparameters - Boxplots
# =========================
for hp in categorical_hyperparameters:
    plt.figure(figsize=(10,6))
    
    combined = pd.concat([
        top_10pct.assign(group='Top 10%'),
        bottom_10pct.assign(group='Bottom 10%')
    ])
    
    combined.boxplot(column='test_accuracy', by=[hp, 'group'], grid=False)
    plt.title(f'Test Accuracy by {hp} (Top 10% vs Bottom 10%)')
    plt.suptitle('')
    plt.xlabel(f'{hp} and Group')
    plt.ylabel('Test Accuracy')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# =========================
# Categorical Hyperparameters - Statistical Tables
# =========================
print("\nCategorical Hyperparameter Distributions:\n")

for hp in categorical_hyperparameters:
    print(f"=== {hp} ===")
    
    print("\nTop 10% Counts:")
    display(top_10pct[hp].value_counts())
    
    print("Top 10% Proportions:")
    display(top_10pct[hp].value_counts(normalize=True))
    
    print("\nBottom 10% Counts:")
    display(bottom_10pct[hp].value_counts())
    
    print("Bottom 10% Proportions:")
    display(bottom_10pct[hp].value_counts(normalize=True))
    
    print("\n" + "="*50 + "\n")

# =========================
# Continuous Hyperparameters - LOWESS Smoothing
# =========================
def plot_lowess_comparison(df_top, df_bottom, hyperparameter, frac=0.5):
    lowess = sm.nonparametric.lowess
    
    z_top = lowess(df_top['test_accuracy'], df_top[hyperparameter], frac=frac)
    z_bottom = lowess(df_bottom['test_accuracy'], df_bottom[hyperparameter], frac=frac)
    
    plt.figure(figsize=(8,5))
    plt.scatter(df_top[hyperparameter], df_top['test_accuracy'], alpha=0.5, label='Top 10% Data', color='blue')
    plt.plot(z_top[:,0], z_top[:,1], 'b-', label='Top 10% LOWESS')
    
    plt.scatter(df_bottom[hyperparameter], df_bottom['test_accuracy'], alpha=0.5, label='Bottom 10% Data', color='red')
    plt.plot(z_bottom[:,0], z_bottom[:,1], 'r-', label='Bottom 10% LOWESS')
    
    plt.title(f'LOWESS Smoothed Test Accuracy vs {hyperparameter}')
    plt.xlabel(hyperparameter)
    plt.ylabel('Test Accuracy')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plot LOWESS for true continuous hyperparameters only
for hp in continuous_hyperparameters:
    if pd.api.types.is_numeric_dtype(plot_data[hp]):
        plot_lowess_comparison(top_10pct, bottom_10pct, hp)
```

```python
import matplotlib.ticker as ticker

def plot_lowess_weight_decay_fine_decimal(df_top, frac=0.5):
    lowess = sm.nonparametric.lowess
    
    z_top = lowess(df_top['test_accuracy'], df_top['weight_decay'], frac=frac)
    
    plt.figure(figsize=(8,5))
    plt.scatter(df_top['weight_decay'], df_top['test_accuracy'], alpha=0.5, color='blue', label='Top 10% Data')
    plt.plot(z_top[:,0], z_top[:,1], 'b-', label='Top 10% LOWESS')
    
    plt.title('LOWESS Smoothed Test Accuracy vs Weight Decay (Top 10%)')
    plt.xlabel('Weight Decay')
    plt.ylabel('Test Accuracy')
    
    plt.xscale('log')  # Fine x-axis scaling (still log, but human-readable)
    plt.gca().xaxis.set_major_formatter(ticker.FormatStrFormatter('%.5f'))  # Decimal style formatting
    
    plt.grid(True, which='both', axis='both')
    plt.legend()
    plt.tight_layout()
    plt.show()
    
# Call it
plot_lowess_weight_decay_fine_decimal(top_10pct)
```

```python
import matplotlib.ticker as ticker

def plot_lowess_weight_decay_linear(df_top, frac=0.5):
    lowess = sm.nonparametric.lowess
    
    z_top = lowess(df_top['test_accuracy'], df_top['lr'], frac=frac)
    
    plt.figure(figsize=(8,5))
    plt.scatter(df_top['weight_decay'], df_top['test_accuracy'], alpha=0.5, color='blue', label='Top 10% Data')
    plt.plot(z_top[:,0], z_top[:,1], 'b-', label='Top 10% LOWESS')
    
    plt.title('LOWESS Smoothed Test Accuracy vs Weight Decay (Top 10%)')
    plt.xlabel('Weight Decay')
    plt.ylabel('Test Accuracy')
    
    # Linear scale
    plt.xscale('linear')
    
    # Optional: set x-limits based on your data range
    min_wd = df_top['weight_decay'].min()
    max_wd = df_top['weight_decay'].max()
    plt.xlim(min_wd * 0.9, max_wd * 1.1)  # Add some margin
    
    # Decimal x-axis format
    plt.gca().xaxis.set_major_formatter(ticker.FormatStrFormatter('%.5f'))
    
    plt.grid(True, which='both', axis='both')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Call it
plot_lowess_weight_decay_linear(top_10pct)
```
